```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
  return 1 / (1+np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
"""
표준편차 히스토그램 그리기
"""

w = np.random.randn(10000, 1000) * 1
print("w.shape", w.shape)

plt.hist(w.flatten(), bins=100, range=(-4, 4), alpha=0.7, color="blue", edgecolor="black")
plt.title("Histogram of Standard Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
TITLE = "표준편차 1 정규분포 가중치 초기값 + Sigmoid 활성화 함수의 분포도"
HIDDEN_LAYER_SIZE = 5

x = np.random.randn(1000, 100)
node_num = 100
activations = {}

for i in range(HIDDEN_LAYER_SIZE):
  if i != 0:
    x = activations[i-1]
  w = np.random.randn(node_num, node_num) * 1
  a = np.dot(x, w)
  z = sigmoid(a)
  activations[i] = z

plt.figure(figsize=(20, 5))
for i, a in activations.items():
  plt.subplot(1, len(activations), i+1) # row, col, index
  plt.title(f"{i+1}-layer", size=14)
  plt.hist(a.flatten(), bins=100, range=(0,1))
plt.suptitle(TITLE, size=18, weight="bold")

```

```python
TITLE = "표준편차 0.01 정규분포 가중치 초기값 + Sigmoid 활성화 함수의 분포도"
HIDDEN_LAYER_SIZE = 5

x = np.random.randn(1000, 100) * 1
node_num = 100
activations = {}

for i in range(HIDDEN_LAYER_SIZE):
  if i != 0:
    x = activations[i-1]
  w = np.random.randn(node_num, node_num) * 0.01 # 매우 작은 표준편차
  a = np.dot(x, w)
  z = sigmoid(a)
  activations[i] = z

plt.figure(figsize=(20, 5))
for i, a in activations.items():
  plt.subplot(1, len(activations), i+1) # row, col, index
  plt.title(f"{i+1}-layer", size=14)
  plt.hist(a.flatten(), bins=100, range=(0,1))
plt.suptitle(TITLE, size=18, weight="bold")
```

```python
TITLE = "Xavier 가중치 초기값 + Sigmoid 활성화 함수의 분포도"
HIDDEN_LAYER_SIZE = 5

x = np.random.randn(1000, 100) * 1
node_num = 100
activations = {}

for i in range(HIDDEN_LAYER_SIZE):
  if i != 0:
    x = activations[i-1]
  previous_node_num = node_num
  w = np.random.randn(node_num, node_num) * np.sqrt(1.0/previous_node_num) # Xavier Weight
  a = np.dot(x, w)
  z = sigmoid(a)
  activations[i] = z

plt.figure(figsize=(20, 5))
for i, a in activations.items():
  plt.subplot(1, len(activations), i+1) # row, col, index
  plt.title(f"{i+1}-layer", size=14)
  plt.hist(a.flatten(), bins=100, range=(0,1))
plt.suptitle(TITLE, size=18, weight="bold")
```

```python
TITLE = "He 가중치 초기값 + ReLU 활성화 함수의 분포도"
HIDDEN_LAYER_SIZE = 5
x = np.random.randn(1000, 100)
node_num = 100
activations = {}

for i in range(HIDDEN_LAYER_SIZE):
  if i != 0:
    x = activations[i-1]
  previous_node_num = node_num
  w = np.random.randn(node_num, node_num) * np.sqrt(2.0 / previous_node_num) # He Weight
  a = np.dot(x, w)
  z = relu(a)
  activations[i] = z

plt.figure(figsize=(20, 5))
for i, a in activations.items():
  plt.subplot(1, len(activations), i+1) # row, col, index
  plt.ylim((0, 1000))
  plt.title(f"{i+1}-layer", size=14)
  plt.hist(a.flatten(), bins=100, range=(0,1))

plt.suptitle(TITLE, size=18, weight="bold")
```
